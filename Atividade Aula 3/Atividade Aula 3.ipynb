```python
import torch
import torch.nn as nn
import torch.optim as optim
import pandas as pd
from torch.utils.data import DataLoader, TensorDataset
```

```python
df = pd.read_csv('student_exam_data_new.csv')

x = torch.FloatTensor(df[['Study Hours', 'Previous Exam Score']].values)
y = torch.FloatTensor(df['Pass/Fail'].values).view(-1, 1)
```

### 1. Criar Dataset e DataLoader

```python
dataset = TensorDataset(x, y)
train_loader = DataLoader(dataset, batch_size=10, shuffle=True)
```

### 2. Definição do Modelo (Regressão Logística Simples)

```python
class LogisticRegressionModel(nn.Module):
    def __init__(self):
        super(LogisticRegressionModel, self).__init__()
        self.linear = nn.Linear(2, 1) # 2 entradas, 1 saída
        
    def forward(self, x):
        return torch.sigmoid(self.linear(x))

model = LogisticRegressionModel()
```

### 3. Função de perda e Otimizador

```python
criterion = nn.BCELoss() # Binary Cross Entropy
optimizer = optim.SGD(model.parameters(), lr=0.01)
```

### 4. Laço de Treinamento

```python
epochs = 100
for epoch in range(epochs):
    for batch_x, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = model(batch_x)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
    
    if (epoch+1) % 20 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')
```
